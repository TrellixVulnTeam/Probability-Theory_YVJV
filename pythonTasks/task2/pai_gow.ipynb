```python
import math
import matplotlib.pyplot as plot

%matplotlib inline

probability_distribution = []
win_prob = []
pW = 0.22
pL = 0.2815 + 0.4985
s = 0.0
steps = [i for i in range(0, 31)]
for step in steps:
    i = pL ** step * pW
    s = s + i
    probability_distribution.append(s)
    win_prob.append(i)
plot.title("Probability distribution")
plot.xlabel('number of games')
plot.ylabel('probability')
plot.plot(steps, probability_distribution)
plot.show()
plot.title("Probability of winning")
plot.xlabel('number of games')
plot.ylabel('probability')
plot.plot(steps, win_prob)
plot.show()
```

```python
vi = 1.96 * 1376.4982
n = 1000
games = []
up = []
low = []
ev = []
for i in range(1, n + 1):
    games.append(i)
    thPayback = vi / math.sqrt(i)
    up.append(thPayback)
    low.append(-1 * thPayback)
    ev.append(20.9)
plot.title("Volatility")
plot.ylabel("Win, % * 100")
plot.xlabel("number of games")
plot.plot(games, up, color='green', linestyle='dashed', label='up')
plot.plot(games, ev, color='blue', label='ev')
plot.plot(games, low, color='red', linestyle='dashed', label='low')
plot.legend()
plot.show()
```

```python
#emperic
f = open("funProb.txt", 'r')
e = []
for t in f.readline().split(" "):
    if t != '':
        e.append(float(t))
plot.title("Empirical function of the probability of winning")
plot.ylabel("probability")
plot.xlabel("number of games")
plot.plot(games[:e.__len__()], e)
plot.show()

prob_e = [e[0]]
for i in range(2, e.__len__()):
    prob_e.append(sum(e[:i]))
plot.title("Empirical probability distribution")
plot.ylabel("probability")
plot.xlabel("number of games")
plot.plot(games[:prob_e.__len__()], prob_e)
plot.show()
```

```python
f = open("thPayback.txt", 'r')
up.clear()
low.clear()
ev.clear()
for thPayback in f.readline().split(" "):
    if thPayback != '':
        up.append(float(thPayback))
        low.append(-1 * float(thPayback))
        ev.append(20.56)

plot.title("Emperical volatility")
plot.ylabel("Win, % * 100")
plot.xlabel("number of games")
plot.plot(games, up, color='green', linestyle='dashed', label='up')
plot.plot(games, ev, color='blue', label='ev')
plot.plot(games, low, color='red', linestyle='dashed', label='low')
plot.legend()
plot.show()
```

```python
f = open("averageTotalWin.txt", 'r')
up.clear()
low.clear()
ev.clear()
for averageTotalWin in f.readline().split(" "):
    if averageTotalWin != '':
        up.append(-1 * float(averageTotalWin))
        low.append(float(averageTotalWin))
        ev.append(20.56)
plot.title("Average total win")
plot.ylabel("Win / loss")
plot.xlabel("number of games")
plot.plot(games[:up.__len__()], up, color='green', linestyle='dashed', label='up')
plot.plot(games[:ev.__len__()], ev, color='blue', label='ev')
plot.plot(games[:low.__len__()], low, color='red', linestyle='dashed', label='low')
plot.legend()
plot.show()
```
